```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from scipy.stats import t
from scipy.stats import ttest_ind
```

```python
# Upload Quantified file
df = pd.read_csv('Quantified.csv')

df.head()
```

```python

# Select the attributes for which you want to compute statistics
attributes = ['Nscore', 'Escore', 'Oscore', 'Ascore', 'Cscore', 'Impulsivity', 'SS']

# Create a categorical variable 'UserType' based on the 'Cannabis' column
df['UserType'] = df['Cannabis'].apply(lambda x: 'Non-User' if x in ['CL0', 'CL1', 'CL2'] else 'User')
```

```python
# Calculate mean values and standard deviations
mean_values = {}
std_devs = {}

# Assuming 'df' is your DataFrame and 'attributes' is a list of attributes
for attribute in attributes:
    for group_name, group_data in df.groupby('UserType'):
        mean_values.setdefault(group_name, {})[attribute] = group_data[attribute].mean()
        std_devs.setdefault(group_name, {})[attribute] = group_data[attribute].std()

# Print mean values
print("Mean Values:")
for group_name in mean_values:
    print(f"  {group_name}:")
    for attribute, mean_value in mean_values[group_name].items():
        print(f"    {attribute}: {mean_value:.5f}")

# Print standard deviations
print("\nStandard Deviations:")
for group_name in std_devs:
    print(f"  {group_name}:")
    for attribute, std_dev in std_devs[group_name].items():
        print(f"    {attribute}: {std_dev:.5f}")
```

```python
# Calculate standard error of the mean (SEM) for each attribute
sem = {}
for attr in attributes:
    for group_name in mean_values:
        sem.setdefault(group_name, {})[attr] = std_devs[group_name][attr] / (len(df[df['UserType'] == group_name]) ** 0.5)

# Print standard errors of the mean
print("Standard Errors of the Mean:")
for group_name in sem:
    print(f"  {group_name}:")
    for attribute, sem_value in sem[group_name].items():
        print(f"    {attribute}: {sem_value:.5f}")
```

```python
# Calculate 95% confidence intervals for mean values
confidence_level = 0.95
degrees_of_freedom = len(df[df['UserType'] == 'User']) - 1  # Assuming equal sample sizes
confidence_intervals = {}
for attr in attributes:
    for group_name in mean_values:
        mean = mean_values[group_name][attr]
        sem_val = sem[group_name][attr]
        confidence_intervals.setdefault(group_name, {})[attr] = t.interval(
            confidence_level, degrees_of_freedom, loc=mean, scale=sem_val
        )
```

```python
# Print mean values, standard deviations, and confidence intervals
for group_name in mean_values:
    print(f"Mean Values for {group_name}:")
    for attribute, mean_value in mean_values[group_name].items():
        print(f"  {attribute}: {mean_value:.5f} ± {std_devs[group_name][attribute]:.4f}")
        print(f"    95% Confidence Interval: {confidence_intervals[group_name][attribute]}")
```

```python

# Initialize dictionary to store p-values
p_values = {}

# Perform t-test for each attribute
for attr in attributes:
    user_data = df[df['UserType'] == 'User'][attr]
    non_user_data = df[df['UserType'] == 'Non-User'][attr]
    t_statistic, p_value = ttest_ind(user_data, non_user_data)
    p_values[attr] = p_value

# Print p-values and determine significance
print("P-values:")
for attribute, p_value in p_values.items():
    print(f"  {attribute}: {p_value:.4f}")
    if p_value < 0.05:
        print(f"    Difference in means for {attribute} is significant (p < 0.05)")
    else:
        print(f"    Difference in means for {attribute} is not significant (p >= 0.05)")
```

```python

# Initialize dictionary to store classification errors
classification_errors = {}

# Iterate over each attribute
for attribute in attributes:
    # Find unique values of the attribute
    unique_values = df[attribute].unique()
    unique_values.sort()

    errors = []
    # Iterate over unique values as potential thresholds
    for threshold in unique_values:
        # Predict class based on the threshold
        predictions = np.where(df[attribute] > threshold, 'User', 'Non-User')
        # Calculate classification error
        error = 1 - accuracy_score(df['UserType'], predictions)
        errors.append((threshold, error))
    
    # Find the threshold with the minimum classification error
    best_threshold, min_error = min(errors, key=lambda x: x[1])
    classification_errors[attribute] = {'best_threshold': best_threshold, 'min_error': min_error}

# Sort attributes based on minimum classification error
sorted_attributes = sorted(classification_errors.keys(), key=lambda x: classification_errors[x]['min_error'])

# Print classification errors and arrange attributes based on prediction ability
print("Classification Errors:")
for attribute in sorted_attributes:
    print(f"{attribute}: Best Threshold = {classification_errors[attribute]['best_threshold']}, "
          f"Min Error = {classification_errors[attribute]['min_error']:.4f}")

# Attribute with the lowest classification error is considered the best predictor
best_predictor = sorted_attributes[0]
print(f"\nBest Predictor: {best_predictor}")
```
